```python
from sklearn.datasets import load_diabetes

from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet 
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
# call for train_test_split model

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
# Apply first 'Linear_Regression' here

reg = LinearRegression()
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)

print("R2_score:", r2_score(y_test,y_pred))
```

```python
# Now, let's look for 'Ridge' on alpha=0.1

reg = Ridge(alpha=0.1)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)

print("R2_score:", r2_score(y_test,y_pred))
```

```python
# Now, try it for 'Lasso' with alpha=0.01

reg = Lasso(alpha=0.01)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)

print("R2_score:", r2_score(y_test,y_pred))
```

```python
# so,we can see here that "Ridge" perform slighlty better than "Lasso" reression
```

```python
# It's time for "ElasticNet_Regression" :-

reg = ElasticNet(alpha=0.005, l1_ratio=0.9)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)

print("R2_score:", r2_score(y_test,y_pred))

# you can alos try for different values of 'alpha' and 'l1_ratio'.
```

```python
# (l1_ratio = 0.9) means that we want 90% 'Ridge part' and 10% 'Lasso part'.
```
